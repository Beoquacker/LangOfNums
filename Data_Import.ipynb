```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = "https://markets.businessinsider.com/news/nvda-stock?p=1"
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
l = a
df = pd.DataFrame(columns=['datetime','title','source','link'])
for el in l:
    df = pd.DataFrame(columns=['datetime','title','source','link'])
    for i in range(1,20):
        url = f"https://markets.businessinsider.com/news/{el}?p={i}"
        response = requests.get(url)
        soup = bs(response.text, 'html.parser')
        articles = soup.find_all('div', class_ = 'latest-news__story')
        for article in articles:
            datetime = article.find('time', class_ = 'latest-news__date').get('datetime')
            title = article.find('a', class_ = 'news-link').text
            link = article.find('a', class_ = 'news-link').get('href')
            source = article.find('span', class_ = 'latest-news__source').text

            row = pd.DataFrame([[datetime,title,source,link]], columns=df.columns)

            df = pd.concat([df,row], ignore_index=True)
    df.to_csv(f'data/{el}.csv')
```
